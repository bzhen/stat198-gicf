## Imports

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Variables

```python
train_steps, train_losses, train_accs = [], [], []
test_steps, test_losses, test_accs = [], [], []
test_next = False
```

## Parsing

```python
with open("cnn_output", 'r') as infile:
    for line in infile:
        if line.startswith("Evaluation:"):
            test_next = True
            continue
        
        if line.startswith("2017"):
            split = line.split()
            step = int(split[2].replace(",", ""))
            loss = float(split[4].replace(",", ""))
            acc = float(split[6].replace(",", ""))

            if test_next:
                test_steps.append(step)
                test_losses.append(loss)
                test_accs.append(acc)
                test_next = False
            else:
                train_steps.append(step)
                train_losses.append(loss)
                train_accs.append(acc)
```

## Plotting

```python
plt.plot(test_accs)
plt.title("Accuracy of test set")
plt.xlabel("Step")
plt.ylabel("Accuracy")
```

```python
plt.plot(test_losses)
plt.title("Loss of Test Set")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.xlim(0,280)
```

```python
plt.plot(train_accs)
plt.title("Accuracy of Train Set")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.ylim(0,2)
plt.xlim(0,28200)
```

```python
plt.plot(train_losses)
plt.title("Loss of Train Set")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.ylim(0,2)
plt.xlim(0,28200)
```
